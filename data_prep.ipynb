```python
import pdfplumber
import pandas as pd

pdf_path = "dataset.pdf"

all_data = []
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        tables = page.extract_tables()
        for table in tables:
            for row in table[1:]:
                all_data.append(row)

columns = ["Date", "Time", "Transaction_ID", "UPI_ID", "Merchant",
           "Amount", "Transaction_Type", "Status", "Remarks"]
df_extracted = pd.DataFrame(all_data, columns=columns)

df_extracted["Amount"] = pd.to_numeric(df_extracted["Amount"], errors='coerce')
```

```python
data = {
    "Date": df_extracted["Date"],
    "Time": df_extracted["Time"],
    "Amount": df_extracted["Amount"],
    "Receiver": df_extracted["Merchant"],
    "Category": df_extracted["Remarks"],
    "Transaction_ID": df_extracted["Transaction_ID"],
    "UPI_ID": df_extracted["UPI_ID"],
    "Transaction_Type": df_extracted["Transaction_Type"],
    "Status": df_extracted["Status"],
}

df = pd.DataFrame(data)

print(df.head())
```

```python
df.to_csv("upi_data.csv", index=False)
```
